```python
from bs4 import BeautifulSoup
import requests
import csv
import html5lib
import pandas as pd
csv_file = open("cms_scrape_imdb.csv", "w", newline="", encoding="utf-8-sig")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Name", "Date", "Rating"])


source = requests.get("https://www.imdb.com/list/ls055592025/").text
soup = BeautifulSoup(source, "lxml")


for element in soup.find_all("div", class_="lister-item mode-detail"):
    
    name = element.h3.a.text.strip()
    
    date = element.h3.find("span", class_="lister-item-year text-muted unbold").text.strip()
    
    rating = element.p.find("span", class_="certificate").text.strip()
    
    csv_writer.writerow([name, date, rating])
    
csv_file.close()


imdb_sheet = pd.read_csv("cms_scrape_imdb.csv")
imdb_sheet
```

```python
from bs4 import BeautifulSoup
import requests
import csv
import html5lib
import pandas as pd

csv_file2 = open("cms_scrape_LOC.csv", "w", newline="", encoding="utf-8")
csv_writer2 = csv.writer(csv_file2)
csv_writer2.writerow(["Title", "Item Desc", "Hyperlink"])

page = 1

source2 = requests.get("https://www.loc.gov/search/?q=cats&sp=1").text
soup2 = BeautifulSoup(source2, "lxml")
while page != 6:
    for element in soup2.find_all("div", class_="description"):
        
        title = element.find("span", class_="item-description-title").a.text.strip()
    
        item_desc = element.find("span", class_="item-description-abstract").text.strip()
    
        hyperlink = element.find("span", class_="item-description-title").a.get("href")
    
        csv_writer2.writerow([title, item_desc, hyperlink])
    
    page += 1
    
    csv_file.close()
    
```
