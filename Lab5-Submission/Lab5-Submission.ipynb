```python
#Scraping https://hoopshype.com/ 
# Importing everything
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium import webdriver


#Scraping Imports
from bs4 import BeautifulSoup
import requests
import csv
import html5lib
import pandas as pd




# Path for the driver
PATH = Service("/Users/edwardsherman\chromedriver.exe")

driver = webdriver.Chrome(service=PATH)

driver.get("https://hoopshype.com/")
print(driver.title)

hoops_link = driver.find_element(By.ID, "menu-item-458969")
hoops_link.click()

#Figured out that you can click on the element you want to access and select copy XPATH
salaries_button = driver.find_element(By.XPATH, "//*/div/div[3]/ul/li[2]/a")
salaries_button.click()


csv_file = open("scrape_nba.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Salary"])

try:
    main = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID,"main"))
    )
    box = main.find_elements(By.TAG_NAME,"tr")
    for i in box:
        salary = i.find_element(By.CLASS_NAME,"hh-salaries-sorted")
        csv_writer.writerow([salary.text])
finally:
    driver.quit()
    csv_file.close()
    
sheet = pd.read_csv("scrape_nba.csv")
sheet
```
