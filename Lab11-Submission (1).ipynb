```python
#Use The Trivia API to create a trivia game with Tkinter
#We’re going to use this API call:
#https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy 
#What are the elements we will need to build to accomplish this? 
#Function that will randomly choose a question from the API call 
#Need a way to tie our radio buttons to the answers – both correct & incorrect 
#Tkinter String Variable 

from tkinter import *
import requests
import json
```

```python
import random
from tkinter import *
import requests

root = Tk()
root.title("Trivia Game")
root.geometry("700x700")

trivia = requests.get("https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy").json()


canvas = Canvas(root, width=800, height=250, bg="khaki1")
canvas.place(relx=.5, rely=.5, anchor=N)

points = 0

def get_trivia():
    global all_answers, correct_answer, que
    all_answers = []
    correct_answer = []
    que = []
    for question in trivia:
        huh = (question["question"])
        que.append(huh)
        answers = question["incorrectAnswers"] + [question["correctAnswer"]]
        all_answers.append(answers)
        correct = question['correctAnswer']
        correct_answer.append(correct)

get_trivia()

def pick_que_ans():
    global current_correct_answer, current_question, current_answers
    x = random.randint(0, 9)
    current_question = que[x]
    current_answers = all_answers[x]
    current_correct_answer = correct_answer[x]
    random.shuffle(current_answers)
    return current_question, current_answers, current_correct_answer

def selection(correct_ans):
    def inner_selection():
        global points
        ans = radio.get()
        if ans == correct_ans:
            label.config(text="You're right!")
            points += 1
        else:
            label.config(text="You're wrong :( !")
        points_label.config(text=f"Points: {points}")
    return inner_selection

def update_question():
    pick_que_ans()
    question_label.config(text=current_question)
    len1 = len(current_answers)
    for num in range(len1):
        ans = current_answers[num]
        radio_buttons[num].config(text=ans, command=selection(current_correct_answer), value=ans)

pick_que_ans()
radio = StringVar()
radio.set("")

question_label = Label(text=current_question, font=("Helvetica"))
question_label.pack()

radio_buttons = []
for ans in current_answers:
    radio_btn = Radiobutton(root, text=ans, variable=radio, value=ans, command=selection(current_correct_answer))
    radio_buttons.append(radio_btn)

for r in radio_buttons:
    r.pack()

label = Label(root)
label.pack()

points_label = Label(root, text=f"Points: {points}")
points_label.pack()

update_button = Button(root, text="Next Question", command=update_question)
update_button.pack()

root.mainloop()
```
