```python
# Path for the driver
PATH = Service("/Users/edwardsherman\chromedriver.exe")

driver = webdriver.Chrome(service=PATH)

driver.get("https://www.youtube.com/@ArielBissett/")
print(driver.title)


click_xpath = "//*[@id='tabsContent']/tp-yt-paper-tab[2]/div"
link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, click_xpath)))
link.click()


csv_file = open("scrape_yt.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Link", "Title", "Views", "Date"])

try:
    main = WebDriverWait(driver, 20).until(
    EC.presence_of_element_located((By.CLASS_NAME,"style-scope ytd-rich-item-renderer"))
    )
#    box = main.find_elements(By.XPATH,"//*[@id='contents']")
# class for this element = style-scope ytd-rich-grid-renderer
#    box = driver.find_elements(By.ID, "contents")
    for i in range(0,10):
        driver.find_element(By.CSS_SELECTOR, "body").send_keys(Keys.CONTROL + Keys.END)
    for x in driver.find_elements(By.CLASS_NAME, "style-scope ytd-rich-item-renderer"):
        #Don't forget the dot ( . ) for looping
        link = x.find_element(By.ID, "video-title-link").get_attribute("href")
        
#        title = x.find_element(By.XPATH, ".//*[@id='video-title']").text        
        title = x.find_element(By.ID, "video-title").text

        views = x.find_element(By.XPATH, ".//*[@id='metadata-line']/span[1]").text
#        views = x.find_elements(By.CLASS_NAME,"inline-metadata-item style-scope ytd-video-meta-block")

        date = x.find_element(By.XPATH, ".//*[@id='metadata-line']/span[2]").text
        print(link, title, views, date)
        
        csv_writer.writerow([link, title, views, date])
finally: 
    driver.quit()
    csv_file.close()
    
yt_sheet = pd.read_csv("scrape_yt.csv")
yt_sheet
```

```python
from turtle import Turtle, Screen
from random import randrange

michelangelo = Turtle()    # 1.  Create two turtles
leonardo = Turtle()
michelangelo.color('orange')
leonardo.color('blue')
michelangelo.shape('turtle')
leonardo.shape('turtle')

michelangelo.up()          # 2.  Pick up the pen so we donâ€™t get lines
leonardo.up()
michelangelo.goto(-100,20)
leonardo.goto(-100,-20)

# Part A -  complete part A here

for x in range(1,10):
  michael = randrange(1,10)
  michelangelo.forward(michael)

  leo = randrange(1,10)
  leonardo.forward(leo)

michelangelo.goto(-100,20)
leonardo.goto(-100,-20)



# Part B - complete part B here

for x in range(1,4):
    num_sides = 3
    leonardo.down()
    leonardo.forward(90)
    leonardo.right(360/num_sides)
leonardo.up()    
leonardo.goto(-100,-120)

for x in range(1,5):
    num_sides2 = 4
    leonardo.down()
    leonardo.forward(90)
    leonardo.right(360,num_sides2)




# The lines below keep the window open after your porgram runs until you click it so you can see what is going on.
window = Screen()
window.exitonclick()
```
