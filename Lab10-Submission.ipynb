```python
#GUI #1 (5pts)
#Using the quotes API, build a GUI that provides a drop-down menu of people randomly chosen from the API
#I want a user to be able to pick from that list and receive a quote that corresponds to that person 
#Ideally, I want to be able to reset that menu & allow people to choose from a new list 

from tkinter import *
import requests
import json
```

```python
import random
root = Tk()
root.title("Random Quote Generator by Author")
root.geometry("500x500")

people_list = []

def get_people():
    global people_list
    data = requests.get("https://api.quotable.io/authors").json()
    data2 = data["results"]
    x = len(data2)
    for i in range(1, x):
        y = random.randint(1, x-1)
        people = data2[y]["name"]
        people_list.append(people)
    


get_people()

def generate_quote():
    person = value.get()
    response = requests.get(f"https://api.quotable.io/quotes?author={person}").json()
    quote = response["results"][0]["content"]
    quote_label.config(text=quote)
    value.set("Select an Author")

def chosen_option():
    author = value.get()
    result_label = Label(root, text=f"You've chosen {author}")
    result_label.pack()

value = StringVar()
value.set("Select an Author")

dropdown_menu = OptionMenu(root, value, *people_list)
choose_button = Button(root, text="Click to choose", command=generate_quote)
quote_label = Label(root, text="", wraplength=400, font=("Helvetica", 14))

dropdown_menu.pack()
choose_button.pack()
quote_label.pack()

root.mainloop()
```
